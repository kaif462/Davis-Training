```python

sales = {}

with open("sales.txt") as f:
    for line in f:
        try:
            p, c, price, qty = line.strip().split(",")
            total = int(price)*int(qty)
            sales[c] = sales.get(c, 0) + total
        except:
            continue

print(sales)
```
